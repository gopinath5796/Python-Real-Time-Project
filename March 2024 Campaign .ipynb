```python
# Create an empty DataFrame with column names and data types:
schema={'Date': 'object', 'Campaign Name': 'object', 'AD Spent': 'float64','No.of Leads': 'int64',
        'Cost Per Lead': 'float64','CTR': 'float64','Link Click': 'int64','LP Views': 'int64',
        'Member Passed': 'float64','LP Conversion %': 'float64'}
output_df = pd.DataFrame(columns=schema.keys()).astype(schema)
output_df
```

```python
# Required Columns for the calculations are stored in the Independent (Variable Name):
independent=dataset[['Day', 'Campaign name', 'Amount spent (INR)', 'Results', 'CTR (link click-through rate)', 'Link clicks', 'Landing page views']]
independent
```

```python
# Creating the Function for Date Conversion, NAN & Infinity Value is Replaced into "0":
def Convert_Date(date_str):
    # Parse the input date string
    date_obj = pd.to_datetime(date_str, format='%m/%d/%Y')
    # Construct the new date format
    return date_obj.strftime("%d %b (%a)")

def NAN_Replace(NAN_Replace):
    if math.isnan(NAN_Replace):
        return 0
    else:
        return NAN_Replace

def INF_Replace(INF_Replace):
    if math.isinf(INF_Replace):
        return 0
    else:
        return INF_Replace
    
```

```python
# Doing Major changes in the independent dataset:
independent['Day'] = independent['Day'].apply(Convert_Date)

independent['Results'] = independent['Results'].apply(NAN_Replace)

independent['CTR (link click-through rate)'] = independent['CTR (link click-through rate)'].apply(NAN_Replace)

independent['Link clicks'] = independent['Link clicks'].apply(NAN_Replace)

independent['Landing page views'] = independent['Landing page views'].apply(NAN_Replace)
```

```python
# Displaying the modified DataFrame:
output_df['Date']=independent['Day']

output_df['Campaign Name']=independent['Campaign name']

output_df['AD Spent']=independent['Amount spent (INR)']

output_df['No.of Leads']=independent['Results']

output_df['Cost Per Lead'] = (output_df['AD Spent'] / output_df['No.of Leads']).apply(lambda x: float('{:.2f}'.format(x)))
output_df['Cost Per Lead'] = output_df['Cost Per Lead'].apply(NAN_Replace)
output_df['Cost Per Lead'] = output_df['Cost Per Lead'].apply(INF_Replace)

output_df['CTR'] = (independent['CTR (link click-through rate)']).apply(lambda x: float('{:.2f}'.format(x)))
output_df['CTR'] = (output_df['CTR'].astype(str) + '%')

output_df['Link Click']=independent['Link clicks']

output_df['LP Views']=independent['Landing page views']

output_df['Member Passed']=((output_df['LP Views'] / output_df['Link Click'])*100).apply(lambda x: float('{:.2f}'.format(x)))
output_df['Member Passed'] = output_df['Member Passed'].apply(NAN_Replace)
output_df['Member Passed'] = (output_df['Member Passed'].astype(str) + '%')

output_df['LP Conversion %']=((output_df['No.of Leads'] / output_df['LP Views'])*100).apply(lambda x: float('{:.2f}'.format(x)))
output_df['LP Conversion %'] = output_df['LP Conversion %'].apply(NAN_Replace)
output_df['LP Conversion %'] = output_df['LP Conversion %'].apply(INF_Replace)
output_df['LP Conversion %'] = (output_df['LP Conversion %'].astype(str) + '%')
```

```python
# Sort the DataFrame by 'Date' column in Ascending order:
output_df= output_df.sort_values(by=['Date'], ascending=[True])

# Save the sorted DataFrame to a CSV file:
output_df.to_csv('output_file_Date_Sorted.csv', index=False)

output_df
```

```python
# Sort the DataFrame by 'Campaign Name' column in Ascending order:
output_df_sorted = output_df.sort_values(by=['Campaign Name'], ascending=[True])

# Save the sorted DataFrame to a CSV file
output_df_sorted.to_csv('output_file_CampaignName_Sorted.csv', index=False)

output_df_sorted
```
